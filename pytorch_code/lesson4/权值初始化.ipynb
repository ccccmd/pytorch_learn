nn.init.calculate_gain  
主要功能：计算激活函数的方差变化尺度  
主要参数  
* nonlinearity:激活函数名称  
* param:激活函数的参数，如Leaky ReLU的negative_slop

```python
import torch
import torch.nn as nn
x = torch.randn(10000)
out = torch.tanh(x)
gain = x.std()/out.std()
gain
```

```python
tanh_gain = nn.init.calculate_gain('tanh')
tanh_gain
```
